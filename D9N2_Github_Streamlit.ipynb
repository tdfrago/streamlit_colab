# Step 1: Create a GitHub Repository


1. Go to GitHub and log in.
2. Click the "+" button in the top right and select New repository.
3. Enter a Repository name (e.g., streamlit_colab).
4. (Optional) Add a description and choose to make it public.
5. Initialize a README file
6. Click Create repository.
7. Copy the repository URL (it should look like https://github.com/[your-username]/streamlit_colab.git)
